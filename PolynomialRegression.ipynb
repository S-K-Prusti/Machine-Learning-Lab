```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Example dataset
X = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5, 7, 8])
```

```python
def polynomial_regression(degree, X, y):
    # Polynomial feature transformation
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)

    # Model training
    model = LinearRegression()
    model.fit(X_poly, y)

    # Prediction
    y_pred = model.predict(X_poly)

    # Error metrics
    mae = mean_absolute_error(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y, y_pred)

    # Print metrics
    print(f"\nDegree {degree} Polynomial Regression")
    print(f"MAE  : {mae:.4f}")
    print(f"MSE  : {mse:.4f}")
    print(f"RMSE : {rmse:.4f}")
    print(f"R²   : {r2:.4f}")

    # Plot
    X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
    X_plot_poly = poly.transform(X_plot)
    y_plot = model.predict(X_plot_poly)

    plt.scatter(X, y, color='red', label='Actual Data')
    plt.plot(X_plot, y_plot, color='blue', label=f'Degree {degree} Fit')
    plt.title(f'Polynomial Regression (Degree {degree})')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.show()
```

```python
polynomial_regression(degree=1, X=X, y=y)
```

```python
polynomial_regression(degree=2, X=X, y=y)
```

```python
polynomial_regression(degree=3, X=X, y=y)
```

```python
polynomial_regression(degree=4, X=X, y=y)
```
